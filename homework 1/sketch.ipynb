## Approximation, Q5

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-5, 5, 100)
```

```python
y_a = relu(x) - relu(-x)
plt.figure(figsize=(3, 3))
plt.plot(x, y_a, label='linear output')
```

```python
y_b = relu(x) + relu(3*x)
plt.figure(figsize=(3, 3))
plt.plot(x, y_b, label='linear output')
```
